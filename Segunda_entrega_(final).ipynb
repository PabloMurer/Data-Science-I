#### III. Exploracion datos faltantes.

Si bien en punto anterior con la sentencia df.info() se menciona no hay datos nulos/faltantes, la realidad es que se esta tomando los valores o resultados NaN como un dato. De todas formas, al realizar un analisis pormenorizado, observamos que las columnas Age, Height, Weight y Medal contienen valores NaN, que seria equivalente a nulo.

Para la columna Medal no es un problema, dado que cada atleta obtiene como resultado una medalla (oro, plata o bronce) o ninguna (NaN).

Pero si constituye un problema para las columnas que deseamos explorar luego, como ser Age, Height, Weight (Edad, altura y peso).

A continuacion realizamos un analisis pormenorizado para entender el comportamiento de los datos faltantes para estas variables:

```python
## A continuacion calcularemos cantidad de filas con algun dato faltante o NaN en las columnas Age, Weight o Height del dataframe df.
## Si hay un dato faltante en 1, 2 o las 3 columnas, el resultado es verdadero (o la fila es contada)

# Paso 1: Calculamos total de filas con resultado = NaN en columnas 'Age', 'Height', o 'Weight'
nan_counts = df[['Year', 'Season', 'Age', 'Height', 'Weight']].copy()
nan_counts['has_nan'] = nan_counts.isna().any(axis=1)
total_nan = nan_counts.groupby(['Year', 'Season'])['has_nan'].sum().reset_index()

# Paso 2: Contamos cantidad de filas por año ('Year') y estacion ('Season')
total_rows = df.groupby(['Year', 'Season']).size().reset_index(name='total_count')

# Paso 3: Combinamos ambos dataframes calculados en pasos 1 y 2.
result = pd.merge(total_nan, total_rows, on=['Year', 'Season'])

# Paso 4: Calculamos porcentaje de filas con resultado Nan sobre el total de filas por año y estacion
result['nan_percentage'] = (result['has_nan'] / result['total_count'])

# Paso 5: Modificamos nombres a las columnas

result = result.rename(columns={'has_nan': 'Tot_filas_con_datos_faltantes',
                                'total_count': 'total_filas',
                                'nan_percentage': '% faltantes'})

# Paso 6: Devolvemos el resultado
result.head(10)
```

Se procede a continuacion a realizar graficos para visualizar el analisis/dataframe "result":

#### IV. Alternativas evaluadas para reemplazo de datos faltantes.

Generamos dataframe **"dataframe_media"** en la cual reemplazamos los valores faltantes de los campos Age (Edad), Height (altura) y Weight (peso) por el promedio de cada campo/columna segun el deporte del atleta que tiene el dato faltante.

Se determino esta metodologia considerando los valores de estos campos son muy particulares de cada deporte y categoria, no siendo apropiado tratar toda la poblacion de datos en su conjunto.

Existen algunas filas (son muy pocas pero luego nos generaran problemas en el mododelado) cuyo deporte no tiene informacion (ejemplo, todos los deportistas del deporte polo no tienen informacion sobre la altura). En ese caso, la media va a ser Nan. Alli se toma el criterio de reemplazar por la media del dataframe.

Se expone a continuacion el codigo:

```python
import pandas as pd
# Calculate the mean of 'Age', 'Height', and 'Weight' for each sport
dataframe_media = df.copy()
sports_mean = dataframe_media.groupby('Sport')[['Age', 'Height', 'Weight']].mean()

# Function to fill NaN values with sport-specific means
def fill_nan_with_sport_mean(row):
    sport = row['Sport']
    for col in ['Age', 'Height', 'Weight']:
        if pd.isna(row[col]):
            if pd.notna(sports_mean.loc[sport, col]):
                row[col] = sports_mean.loc[sport, col]
            else:
                # If sport mean is NaN, fill with the overall mean
                row[col] = dataframe_media[col].mean()
    return row

# Apply the function to replace NaN values in the dataframe
dataframe_media = dataframe_media.apply(fill_nan_with_sport_mean, axis=1)

dataframe_media.head(10)
```

Procedemos ahora a generar dataframe **"dataframe_moda"** en la cual reemplazamos los valores faltantes de los campos Age (Edad), Height (altura) y Weight (peso) por el la moda de cada campo/columna segun el deporte del atleta que tiene el dato faltante.

Se expone a continuacion el codigo:

```python
import pandas as pd
import numpy as np
# Calculate the mode of 'Age', 'Height', and 'Weight' for each sport
dataframe_moda = df.copy()
sports_mode = dataframe_moda.groupby('Sport')[['Age', 'Height', 'Weight']].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)

# Function to fill NaN values with sport-specific modes
def fill_nan_with_sport_mode(row):
    sport = row['Sport']
    for col in ['Age', 'Height', 'Weight']:
        if pd.isna(row[col]):
            if pd.notna(sports_mode.loc[sport, col]):
                row[col] = sports_mode.loc[sport, col]
            else:
                # If sport mode is NaN, fill with the overall mode
                row[col] = dataframe_moda[col].mode().iloc[0]
    return row

# Apply the function to replace NaN values in the dataframe
dataframe_moda = dataframe_moda.apply(fill_nan_with_sport_mode, axis=1)

dataframe_moda.head(10)
```

```python
dataframe_moda.isnull().sum()
```

Procedemos por ultimo a generar dataframe **"dataframe_mediana"** en la cual reemplazamos los valores faltantes de los campos Age (Edad), Height (altura) y Weight (peso) por la mediana de cada campo/columna segun el deporte del atleta que tiene el dato faltante.

Se expone a continuacion el codigo:

```python
import pandas as pd
# Calculate the median of 'Age', 'Height', and 'Weight' for each sport
dataframe_mediana = df.copy()
sports_median = dataframe_mediana.groupby('Sport')[['Age', 'Height', 'Weight']].median()

# Function to fill NaN values with sport-specific medians
def fill_nan_with_sport_median(row):
    sport = row['Sport']
    for col in ['Age', 'Height', 'Weight']:
        if pd.isna(row[col]):
            if pd.notna(sports_median.loc[sport, col]):
                row[col] = sports_median.loc[sport, col]
            else:
                # If sport median is NaN, fill with the overall median
                row[col] = dataframe_mediana[col].median()
    return row

# Apply the function to replace NaN values in the dataframe
dataframe_mediana = dataframe_mediana.apply(fill_nan_with_sport_median, axis=1)

dataframe_mediana.head(10)
```

A continuacion se calcula la cantidad de outliers para los valores edad, peso y altura:

```python
# Calculate Q1, Q3, IQR, and outlier boundaries for 'Age', 'Weight', and 'Height'
def calculate_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return Q1, Q3, IQR, lower_bound, upper_bound, outliers


columns = ['Age', 'Weight', 'Height']
for col in columns:
    Q1, Q3, IQR, lower_bound, upper_bound, outliers = calculate_outliers(dataframe_moda, col)

    outlier_percentage = (len(outliers) / len(dataframe_moda)) * 100
    print(f"Analysis for {col}:")
    print(f"  Q1: {Q1}")
    print(f"  Q3: {Q3}")
    print(f"  IQR: {IQR}")
    print(f"  Lower Bound: {lower_bound}")
    print(f"  Upper Bound: {upper_bound}")
    print(f"  Number of outliers: {len(outliers)}")
    print(f"  Percentage of outliers: {outlier_percentage:.2f}%")
    print("-" * 20)
```

Visualizamos entonces que el porcentaje de outliers para las variables de peso y altura es relativamente bajo, siendo la variable mas representativa con outliers para la edad de los deportistas (cuyo porcentaje ronda el 4.61%).

Contamos a continuacion la cantidad de atletas (Id's no repetidos) con edad, peso o altura cuyo valor constituye un outlier:

```python
def calculate_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

columns = ['Age', 'Weight', 'Height']
outliers_data = {}

for col in columns:
    outliers = calculate_outliers(dataframe_moda, col)
    outliers_data[col] = outliers

# Count the number of unique IDs with outliers in any of the specified columns
total_ids_with_outliers = 0
for col in columns:
    total_ids_with_outliers += len(outliers_data[col]['ID'].unique())
total_ids_with_outliers = len(set([item for sublist in [outliers_data[col]['ID'].unique() for col in columns] for item in sublist]))

# Calculate the total number of unique IDs
total_unique_ids = len(dataframe_moda['ID'].unique())

# Calculate the percentage
percentage_outliers = (total_ids_with_outliers / total_unique_ids) * 100

print(f"Total IDs with outliers in 'Age', 'Weight', or 'Height': {total_ids_with_outliers}")
print(f"Total unique IDs: {total_unique_ids}")
print(f"Percentage of IDs with outliers: {percentage_outliers:.2f}%")
```

```python
import pandas as pd

def calculate_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

columns = ['Age', 'Weight', 'Height']
outliers_by_sport = {}

for col in columns:
    outliers = calculate_outliers(dataframe_moda, col)
    # Group outliers by sport and count unique IDs
    outliers_by_sport[col] = outliers.groupby('Sport')['ID'].nunique().reset_index(name='Outlier_Count')

# Combine outlier counts for all columns
age_outliers = outliers_by_sport['Age']
weight_outliers = outliers_by_sport['Weight']
height_outliers = outliers_by_sport['Height']

combined_outliers = pd.merge(age_outliers, weight_outliers, on='Sport', how='outer').fillna(0)
combined_outliers = pd.merge(combined_outliers, height_outliers, on='Sport', how='outer').fillna(0)

combined_outliers['Total_Outliers'] = combined_outliers['Outlier_Count_x'] + combined_outliers['Outlier_Count_y'] + combined_outliers['Outlier_Count']
combined_outliers = combined_outliers.sort_values('Total_Outliers', ascending=False)

# Calculate Pareto
combined_outliers['Cumulative_Percentage'] = (combined_outliers['Total_Outliers'].cumsum() / combined_outliers['Total_Outliers'].sum()) * 100

# Show data up to 80% cumulative percentage
pareto_table = combined_outliers[combined_outliers['Cumulative_Percentage'] <= 80]

#Add index column
pareto_table.insert(0, 'Index', range(1, 1 + len(pareto_table)))

print(pareto_table[['Index', 'Sport', 'Total_Outliers', 'Cumulative_Percentage']])
```

##### V.II Outliers: Tratamiento.

Generaremos un nuevo dataframe **"dataframe_no_outliers"** que:

- Reemplace los valores mayores al valor upper bound (bigote superior) por el valor del upper bound. Tanto en las variables edad como peso, la mayoria de outliers estan por "encima" del "bigote superior".
- Reemplace los valores menores al valor lower bound (bigote inferior) por el valor del lower bound.
- Elimine el deporte art competition. Se trata de un deporte que genera muchos outliers, que se dejo de practicar en 1948 y que en realidad era una competencia de arte donde se colocaba la edad del artista original. No constituye informacion valiosa para el analisis que se pretende realizar.

```python
import pandas as pd

def replace_outliers_with_bounds(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_copy = df.copy()  # Create a copy to avoid modifying the original

    df_copy.loc[df_copy[column] < lower_bound, column] = lower_bound
    df_copy.loc[df_copy[column] > upper_bound, column] = upper_bound

    return df_copy


# Apply the function to the 'Age', 'Weight', and 'Height' columns
dataframe_no_outliers = replace_outliers_with_bounds(dataframe_moda, 'Age')
dataframe_no_outliers = replace_outliers_with_bounds(dataframe_no_outliers, 'Weight')
dataframe_no_outliers = replace_outliers_with_bounds(dataframe_no_outliers, 'Height')


# Remove rows where 'Sport' is 'Art Competitions'
dataframe_no_outliers = dataframe_no_outliers[dataframe_no_outliers['Sport'] != 'Art Competitions']

# Now 'dataframe_no_outliers' contains the modified data
dataframe_no_outliers.head(10)
```
